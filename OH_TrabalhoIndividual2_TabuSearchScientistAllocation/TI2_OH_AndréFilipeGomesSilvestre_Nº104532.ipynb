## Resolução do Algoritmo em Python

```python
# Import das Bibliotecas
import pandas as pd
pd.set_option('display.max_colwidth', None)  # Visualizar a informação toda do DataFrame - Será útil mais adiante
```

```python
# Criação do DataFrame com os valores de aptidão de cada cientista para liderar cada um dos projetos
cientistas = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10']
projetos = ['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10']
aptidao = {
    'C1': [70, 65, 55, 50, 90, 67, 80, 62, 100, 85],
    'C2': [74, 83, 54, 60, 100, 75, 50, 76, 82, 45],
    'C3': [71, 87, 66, 58, 74, 81, 48, 52, 100, 64],
    'C4': [50, 89, 78, 63, 51, 40, 48, 100, 68, 75],
    'C5': [100, 66, 83, 77, 54, 58, 93, 89, 53, 48],
    'C6': [80, 55, 70, 65, 94, 47, 60, 88, 73, 60],
    'C7': [87, 63, 90, 79, 47, 77, 90, 76, 85, 90],
    'C8': [67, 95, 100, 40, 70, 54, 70, 100, 65, 57],
    'C9': [90, 45, 88, 48, 65, 68, 80, 46, 71, 67],
    'C10': [67, 77, 50, 60, 100, 70, 80, 60, 65, 70]
}

df_aptidao = pd.DataFrame(aptidao, index=projetos, columns=cientistas).T
df_aptidao
```

```python
# Solução Admissível para o problema
solucao_inicial = {
    'P1': 'C5',
    'P2': 'C3',
    'P3': 'C8',
    'P4': 'C6',
    'P5': 'C2',
    'P6': 'C10',
    'P7': 'C7',
    'P8': 'C4',
    'P9': 'C1',
    'P10': 'C9'
}

# Função que calcula a aptidão total das afetação de cientistas-projetos
def calcular_aptidao_total(solucao):
    aptidao_total = 0
    for projeto, coordenador in solucao.items():
        aptidao_total += df_aptidao.loc[coordenador, projeto]
    return aptidao_total

# Valor de Aptidão Total da Solução Admissível Inicial
print('Aptidão Total da S.A.I.:',calcular_aptidao_total(solucao_inicial))
```

```python
# Definição dos Critérios de Paragem do Algoritmo da Lista Tabu Implementado
# num_max_iteracoes = 100  # Tal como estipulado no Enunciado (alínea i))
num_max_iteracoes = 500  # Considerei um valor superior para tentar encontrar um valor mais 'promissor'
# aptidao_minima = 850 # Tal como estipulado no Enunciado (alínea i))
aptidao_minima = 1000 # Considerei o máximo que hipoteticamente seria possível obter caso todos os Projetos tivessem Aptidão 100

# Função para Gerar a Vizinhança - Retorna uma lista com todos os casos possíveis apenas trocando 2 cientistas.
def gerar_vizinhanca(solucao_atual):
    vizinhanca = []
    for projeto in solucao_atual:
        for projeto_1 in solucao_atual:
            if projeto != projeto_1:
                nova_solution = solucao_atual.copy()
                nova_solution[projeto], nova_solution[projeto_1] = nova_solution[projeto_1], nova_solution[projeto]
                if nova_solution not in vizinhanca:
                    vizinhanca.append(nova_solution)
    return vizinhanca

# Função para encontrar o Melhor Vizinho dentro do Conjunto da Vizinhança
def encontrar_melhor_vizinho(vizinhanca, lista_tabu, solucao_anterior):
    melhor_vizinho = None
    melhor_aptidao = float('-inf')
    
    # Percorre todos as soluções da vizinhança
    for vizinho in vizinhanca:
        vizinho_trocas = []
        
        # Guarda as *Trocas* que ocorreram 
        for projeto, coordenador_atual in vizinho.items():
            coordenador_anterior = solucao_anterior[projeto]
            if coordenador_atual != coordenador_anterior:
                troca = (projeto, coordenador_atual)
                vizinho_trocas.append(troca)
        
        # Verifica se as *Trocas* não estão na Lista Tabu
        if not any(troca in lista_tabu for troca in vizinho_trocas):
            aptidao_vizinho = calcular_aptidao_total(vizinho)
            
            # Atualiza a melhor solução encontrada na vizinhança
            if aptidao_vizinho > melhor_aptidao:
                melhor_vizinho = vizinho
                melhor_aptidao = aptidao_vizinho
            
    return melhor_vizinho, melhor_aptidao

# Função que implementa o Algortimo de Pesquisa Tabu - Recebe como Input uma Solução Admissível Inicial 
def pesquisa_tabu(solucao_inicial):
    # Inicialização
    melhor_solucao = solucao_inicial.copy()
    melhor_aptidao = calcular_aptidao_total(melhor_solucao)
    solucao_atual = melhor_solucao.copy()
    aptidao_atual = melhor_aptidao

    # Executar Algoritmo de Pesquisa Tabu
    iteracao = 0
    lista_tabu = []
    solucoes_iteradas = []
    
    # Definir os Critérios de Paragem do Algoritmo
    while iteracao < num_max_iteracoes and aptidao_atual < aptidao_minima:
        
        # Guardar a solução anterior
        solucao_anterior = solucao_atual
        
        # Gerar a Vizinhança e o Melhor Vizinho tendo em conta a Lista Tabu
        vizinhanca = gerar_vizinhanca(solucao_atual)
        melhor_vizinho, aptidao_vizinho = encontrar_melhor_vizinho(vizinhanca, lista_tabu, solucao_anterior)
        
        # Atualizar a Solução Atual
        solucao_atual = melhor_vizinho
        aptidao_atual = aptidao_vizinho
        
        # Verificar se a Solução Atual é melhor do que a Melhor Solução Encontrada
        if aptidao_atual > melhor_aptidao:
            melhor_solucao = solucao_atual
            melhor_aptidao = aptidao_atual
        
        # Atualizar a Lista Tabu - Remover o 1º e 2º elementos caso o tamanho da lista seja superior a 12 
        #                          (valor escolhido após diversas tentativas)
        if len(lista_tabu) > 12:
            lista_tabu.pop(0)
            lista_tabu.pop(1)
        
        # Atualizar a Lista Tabu - Adicionar apenas as trocas de cientistas que deram origem à melhor solução vizinha
        for projeto, coordenador_atual in solucao_atual.items():
            coordenador_anterior = solucao_anterior[projeto]
            if coordenador_atual != coordenador_anterior:
                troca = (projeto, coordenador_atual)
                lista_tabu.append(troca)
        
        # Registar a Solução Iterada
        solucoes_iteradas.append((str(solucao_anterior),str(solucao_atual), aptidao_atual, lista_tabu.copy()))
        iteracao += 1
    
    # Criação de um DataFrame com todas as Soluções Iteradas no Algoritmo e respetivas Aptidões Totais e Listas Tabu
    df_solucoes_iteradas = pd.DataFrame(solucoes_iteradas, columns=['Solução Anterior','Solução Atual', 'Aptidão Total', 'Lista Tabu']).rename_axis('Iteração')

    return melhor_solucao, melhor_aptidao, df_solucoes_iteradas

# Executar o Algoritmo de Pesquisa Tabu com a Solução Admissível Inicial (S.A.I.) definida anteriormente
melhor_solucao, melhor_aptidao, df_solucoes_iteradas = pesquisa_tabu(solucao_inicial)

# Print dos Resultados Obtidos
print("Melhor Solução encontrada:")
for projeto, coordenador in melhor_solucao.items():
    print(f"{projeto}: {coordenador}")
print("\nAptidão Total:", melhor_aptidao, '\n')
print("\nSoluções Iteradas Distintas (ordenadas pela Aptidão Total):")
df_solucoes_iteradas
```
